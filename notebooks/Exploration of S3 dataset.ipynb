```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python


df = (spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .load("s3://aws-data-lake-databricks/dataset")
)

(df.writeStream
    .format("delta")
    .option("checkpointLocation", "/tmp/checkpoints/spotify_bronze")
    .trigger(availableNow=True)  # 👈 modo batch
    .toTable("dev_workspace.bronze.raw_spotify_tracks")
)
```
