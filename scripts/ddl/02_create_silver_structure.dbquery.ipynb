```sql
-- Crear el schema bronze dentro del catálogo
CREATE SCHEMA IF NOT EXISTS dev_workspace.silver;


--  Crear la tabla silver de spotify tracks
CREATE TABLE IF NOT EXISTS dev_workspace.silver.spotify_tracks (
    track_id            STRING      NOT NULL,
    track_name          STRING      NOT NULL,
    artists             ARRAY<STRING>,               -- parsed from comma-separated string
    album_name          STRING,
    track_genre         STRING,
    popularity          INT,
    duration_ms         BIGINT,
    explicit            BOOLEAN,
    danceability        DOUBLE,
    energy              DOUBLE,
    key                 STRING,                      -- mapped: 0 → 'C', 1 → 'C#', etc.
    loudness            DOUBLE,
    mode                STRING,                      -- mapped: 0 → 'Minor', 1 → 'Major'
    speechiness         DOUBLE,
    acousticness        DOUBLE,
    instrumentalness    DOUBLE,
    liveness            DOUBLE,
    valence             DOUBLE,
    tempo               DOUBLE,
    time_signature      INT,
    _ingested_at        TIMESTAMP,
    _source_file        STRING
)
USING DELTA;

ALTER TABLE dev_workspace.silver.spotify_tracks
    ADD CONSTRAINT chk_popularity
    CHECK (popularity IS NULL OR (popularity >= 0 AND popularity <= 100));

 
ALTER TABLE dev_workspace.silver.spotify_tracks
    ADD CONSTRAINT chk_duration
    CHECK (duration_ms IS NULL OR duration_ms > 0);

 
ALTER TABLE dev_workspace.silver.spotify_tracks
    ADD CONSTRAINT chk_tempo
    CHECK (tempo IS NULL OR tempo > 0);

 
ALTER TABLE dev_workspace.silver.spotify_tracks
    ADD CONSTRAINT chk_danceability
    CHECK (danceability IS NULL OR (danceability >= 0.0 AND danceability <= 1.0));

 
ALTER TABLE dev_workspace.silver.spotify_tracks
    ADD CONSTRAINT chk_energy
    CHECK (energy IS NULL OR (energy >= 0.0 AND energy <= 1.0));
```
